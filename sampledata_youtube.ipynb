```python
print(84 * 14)
font_images = os.listdir("/content/drive/MyDrive/fonts/32_font") 
print(len(font_images))
```

```python
print(font_images)
fontpath = "/content/drive/MyDrive/fonts/32_font/"
```

```python
X_csv = []

for imagename in font_images:
  img_path = os.path.join(fontpath, imagename)
  
  img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)

  left_image = rotate(img, angle = 15, cval = 1)
  right_image = rotate(img, angle = -15, cval = 1)
  
  hori_transform = AffineTransform(translation = (27,0))
  warp_h_image = warp(img, hori_transform, mode="wrap")
  
  verti_transform = AffineTransform(translation = (0,6))
  warp_v_image = warp(img, verti_transform, mode="wrap")

  X_csv.append(img.flatten())
  X_csv.append(left_image.flatten()* 255)
  X_csv.append(right_image.flatten()* 255)
  X_csv.append(warp_h_image.flatten()* 255)
  X_csv.append(warp_v_image.flatten()* 255)
  
```

```python
X_csv = np.array(X_csv)
```
